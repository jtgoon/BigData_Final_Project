```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm
from datetime import datetime

from matplotlib import rc
rc('font',family='AppleGothic')

import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
from sklearn import linear_model
from scipy import stats
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
import glob
import os

import time
import re
```

```python
final_data = pd.read_csv('/Users/hanjaemin/Desktop/final_project/BigData_Final_Project/data/Merge_Data/merge_data.csv')
final_data
```

```python
### 결측행 제거
final_data.dropna(inplace = True)

### 인덱스 재설정
final_data.reset_index(inplace = True, drop = True)
```

```python
final_data
```

```python
# openingdate type 변환(object -> DateTime)
final_data['openingdate'] = pd.to_datetime(final_data['openingdate'])
final_data.info()
```

```python
### 범주변수 숫자화

## country == 한국 : 0 , country !== 한국 : 1
country_list=[]
for i in range(len(final_data['country'])):
    if final_data['country'][i] == '한국':
        country_list.append(0)
    else:
        country_list.append(1)

final_data['country'] = country_list
```

```python
## actor를 이름에서 actor의 수로 변환

final_data['actor'] = final_data.actor.str.count(',') + 1
```

```python
## filmrank 

## 전체관람가 : 0, 12세관람가 : 1, 15세관람가 : 2, 청소년관람불가 : 3, 그 외 : 4


## 연령대만 추출 후 숫자형으로 변환

filmrank_num_list=[]

final_data["filmrank"] = [filmrank_num.split()[1] for filmrank_num in final_data['filmrank']]

for i in range(len(final_data['filmrank'])):
    
    if final_data['filmrank'][i] == '전체':
        filmrank_num_list.append(0)
        
    elif final_data['filmrank'][i] == '12세':
        filmrank_num_list.append(1)
    
    elif final_data['filmrank'][i] == '15세':
        filmrank_num_list.append(2)
        
    elif final_data['filmrank'][i] == '청소년':
        filmrank_num_list.append(3)
        
    else:
        filmrank_num_list.append(4)
        

final_data['filmrank'] = filmrank_num_list
```

```python
## genre 

## 코미디 : 0, 다큐멘터리 : 1, 액션 : 2, 판타지 : 3, 드라마 : 4, 범죄 : 5, 애니메이션 : 6, SF : 7, 모험 :8,
## 미스터리 : 9, 전쟁 : 10, 멜로/로맨스 : 11, 스릴러, 서스펜스 : 12, 서사 : 13, 공포 : 14, 느와르 : 15, 뮤지컬 : 16,
## 가족 : 17, 그 외 : 18


genre_list=[]

for i in range(len(final_data['genre'])):
    
    if final_data['genre'][i] == '코미디':
        genre_list.append(0)
        
    elif final_data['genre'][i] == '다큐멘터리':
        genre_list.append(1)
    
    elif final_data['genre'][i] == '액션':
        genre_list.append(2)
        
    elif final_data['genre'][i] == '판타지':
        genre_list.append(3)
        
    elif final_data['genre'][i] == '드라마':
        genre_list.append(4)
    
    elif final_data['genre'][i] == '범죄':
        genre_list.append(5)
        
    elif final_data['genre'][i] == '애니메이션':
        genre_list.append(6)
        
    elif final_data['genre'][i] == 'SF':
        genre_list.append(7)
    
    elif final_data['genre'][i] == '모험':
        genre_list.append(8)
        
    elif final_data['genre'][i] == '미스터리':
        genre_list.append(9)      
        
    elif final_data['genre'][i] == '전쟁':
        genre_list.append(10)
    
    elif final_data['genre'][i] == '멜로/로맨스':
        genre_list.append(11)
        
    elif final_data['genre'][i] == '스릴러' or final_data['genre'][i] == '서스펜스' :
        genre_list.append(12)

    elif final_data['genre'][i] == '서사':
        genre_list.append(13)
        
    elif final_data['genre'][i] == '공포':
        genre_list.append(14)
        
    elif final_data['genre'][i] == '느와르':
        genre_list.append(15)
        
    elif final_data['genre'][i] == '뮤지컬':
        genre_list.append(16)
        
    elif final_data['genre'][i] == '가족':
        genre_list.append(17)
    
    else:
        genre_list.append(18)
        

final_data['genre'] = genre_list
```

```python
## 영화 배급사

# 국내 메이져 영화사 : CJ E & M, 롯데, 쇼박스, 넥스트엔터테인먼월드(NEW), # 해외 메이져 영화사 : 월트디즈니, 유니버설픽쳐스, 워너스브러더스  -> 0
# 그외 배급사 -> 1

distributor_list=[]

for i in range(len(final_data['distributor'])):
    if final_data['distributor'][i].find('씨제이이엔엠') >= 0 or final_data['distributor'][i].find('CJ ENM') >= 0 or final_data['distributor'][i].find('롯데') >= 0 or final_data['distributor'][i].find('쇼박스') >= 0 or final_data['distributor'][i].find('(NEW)') >=0 or final_data['distributor'][i].find('월트디즈니') >= 0 or final_data['distributor'][i].find('유니') >=0 or final_data['distributor'][i].find('워너브러더스') >= 0:
        distributor_list.append(0)
    else:
        distributor_list.append(1)
    
final_data['distributor'] = distributor_list
```
