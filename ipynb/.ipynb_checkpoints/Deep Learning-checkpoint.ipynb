```python
movie = pd.read_csv('../data/Preprocessing_movie_data/movie.csv', parse_dates=['openingdate'])
movie_one_hot = pd.read_csv('../data/Preprocessing_movie_data/final_data.csv', parse_dates=['openingdate'])
```

### 2. Regression

```python
movie_one_hote_hot.columns
```

```python
x = movie_one_hot.drop(columns=['moviename', 'openingdate', 'sales'])
y = movie_one_hot['sales']
y = y.to_numpy()
```

```python
movie.columns
```

```python
x = movie[['attendance', 'screen', 'playcount', 'director']]
y = movie['sales'].to_numpy()
```

```python
sc = MinMaxScaler()
sc.fit(x)
x_scale = sc.transform(x)
x = pd.DataFrame(data=x_scale, columns=['attendance', 'screen', 'playcount', 'director'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
def build_model():
    model = keras.Sequential([
        layers.Dense(128, activation='relu', input_shape=(4,)),
        layers.Dense(128, activation='relu'),
        layers.Dense(1)
    ])
    
    optimizer = tf.keras.optimizers.Adam(0.1)
    
    model.compile(loss='mse',
                 optimizer=optimizer)
    
    return model
```

```python
model = build_model()
model.summary()
```

```python
epochs = 2000

early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(
    x_train, y_train,
    epochs=epochs, validation_split=0.2, verbose=0,
    callbacks=[early_stop]
)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
plt.figure(figsize=(8, 12))

plt.subplot(2, 1, 1)
plt.xlabel('epoch')
plt.ylabel('mae')
plt.plot(hist['epoch'], hist['mae'], label='train error')
plt.plot(hist['epoch'], hist['val_mae'], label='val error')
plt.legend()

plt.subplot(2, 1, 2)
plt.xlabel('epoch')
plt.ylabel('mse')
plt.plot(hist['epoch'], hist['mse'], label='train error')
plt.plot(hist['epoch'], hist['val_mse'], label='val error')
plt.legend()
```

```python
test_predictions = model.predict(x_test).flatten()

plt.scatter(y_test, test_predictions)
plt.xlabel('True Values [MPG]')
plt.ylabel('Predictions [MPG]')
plt.axis('equal')
plt.axis('square')
plt.xlim([0,plt.xlim()[1]])
plt.ylim([0,plt.ylim()[1]])
_ = plt.plot([-100, 100], [-100, 100])
```

```python
error = test_predictions - y_test
plt.hist(error, bins = 25)
plt.xlabel("Prediction Error [MPG]")
_ = plt.ylabel("Count")
```

## 1. 데이터 불러오기

```python
movie_budget_domestic = pd.read_csv('../data/Budget_Data/Preprocessing_Data/Korea_Data.csv')
movie_budget_abroad = pd.read_csv('../data/Budget_Data/Preprocessing_Data/International_Data.csv')
```

```python
movie_budget_domestic
```

```python
movie_budget_abroad
```

```python
movie_budget_domestic = movie_budget_domestic.drop(columns=['attendance'])
```

```python
movie_budget_domestic.head()
```

```python
movie_budget_abroad = movie_budget_abroad[movie_budget_abroad['sales'] != 0]
```

```python
movie_budget = pd.concat([movie_budget_domestic, movie_budget_abroad], axis=0)
movie_budget
```

```python
movie_budget.reset_index(inplace=True, drop=True)
movie_budget
```

```python
x = movie_budget['sales']
x = x.to_numpy()
y = movie_budget['budget']
y = y.to_numpy()
```

```python
sc = MinMaxScaler()

x = x.reshape(-1, 1)
sc.fit(x)
x = sc.transform(x).reshape(-1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
models = []
models.aeppend(())
```

```python
model = keras.models.Sequential()

model.add(keras.layers.Dense(1, input_dim=1))
model.compile(loss='mse', optimizer=keras.optimizers.SGD(learning_rate=0.01))
result = model.fit(x_train, y_train, epochs=2000, verbose=1)
```
